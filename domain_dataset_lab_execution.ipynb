```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["LANGSMITH_TRACING"]  = "true"
os.environ["LANGSMITH_ENDPOINT"] = "https://eu.api.smith.langchain.com"
os.environ["LANGSMITH_PROJECT"]  = "ai-consulting-qa-eval"
```

```python
from langsmith import Client
from langsmith.wrappers import wrap_openai
from langsmith import traceable
from openai import OpenAI
from openevals.llm import create_llm_as_judge
from openevals.prompts import CORRECTNESS_PROMPT

client = Client()
openai_client = wrap_openai(OpenAI())

print("Connected:", client.list_datasets())
```

```python
DATASET_NAME = "ai-consulting-qa-eval"

existing = [d.name for d in client.list_datasets()]
if DATASET_NAME not in existing:
    dataset = client.create_dataset(
        DATASET_NAME,
        description="Realistic AI consulting questions from SME and marketing clients"
    )
    client.create_examples(
        inputs  = [{"question": e["question"]} for e in examples],
        outputs = [{"answer":   e["answer"]}   for e in examples],
        dataset_id = dataset.id,
    )
    print(f"Created dataset with {len(examples)} examples")
else:
    print(f"Dataset '{DATASET_NAME}' already exists — skipping upload")
```

```python
models = openai_client.models.list()
gpt_models = [m.id for m in models if "gpt-4.1" in m.id]
print(gpt_models)
```

```python
SYSTEM_PROMPT = """You are a friendly and pragmatic AI consultant advising business owners and marketing teams. 

Your communication style:
- Friendly, direct, and never sales-sy — you are on the client's side
- Simple and jargon-free — any client regardless of technical background should understand you
- Concise but complete — no unnecessary padding

For every answer you must:
1. Tell the client upfront whether they actually need a specialised AI tool or whether a premium LLM subscription (like ChatGPT Plus or Claude Pro) would already solve their problem
2. Give a practical overview of 2-3 realistic tools or approaches that fit their situation
3. Be honest about limitations immediately — do not oversell
4. Tell them what they need to have in place before they can move forward (data access, technical resources, budget, integrations)
5. Give a rough sense of implementation cost and time so they can set realistic expectations

What to avoid:
- Generic answers that could apply to anyone
- Technical jargon without explanation
- Recommending complex solutions when a simple one exists
- Overpromising what AI can do
- Never open with phrases like 'Good question!' or 'Great question!' — get straight to the answer
"""
```

```python
@traceable(name="target-gpt4.1-mini")
def target_mini(inputs: dict) -> dict:
    response = openai_client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user",   "content": inputs["question"]},
        ],
        max_tokens=600,
    )
    return {"answer": response.choices[0].message.content.strip()}


@traceable(name="target-gpt4.1-nano")
def target_nano(inputs: dict) -> dict:
    response = openai_client.chat.completions.create(
        model="gpt-4.1-nano",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user",   "content": inputs["question"]},
        ],
        max_tokens=600,
    )
    return {"answer": response.choices[0].message.content.strip()}
```

```python
test_input = {"question": examples[0]["question"]}

print("=== GPT-4.1-mini ===")
print(target_mini(test_input)["answer"])
print("\n=== GPT-4.1-nano ===")
print(target_nano(test_input)["answer"])
```

```python
from openevals.llm import create_llm_as_judge
from openevals.prompts import CORRECTNESS_PROMPT

_judge = create_llm_as_judge(
    prompt=CORRECTNESS_PROMPT,
    model="openai:gpt-4.1-mini",
    feedback_key="correctness",
)

def correctness_evaluator(inputs: dict, outputs: dict, reference_outputs: dict) -> dict:
    return _judge(
        inputs=inputs,
        outputs=outputs,
        reference_outputs=reference_outputs
    )
```

```python
test_output = target_mini({"question": examples[0]["question"]})

test_eval = correctness_evaluator(
    inputs={"question": examples[0]["question"]},
    outputs=test_output,
    reference_outputs={"answer": examples[0]["answer"]}
)

print(test_eval)
```

```python
results_mini = client.evaluate(
    target_mini,
    data=DATASET_NAME,
    evaluators=[correctness_evaluator],
    experiment_prefix="gpt-4.1-mini",
    max_concurrency=2,
)
print(results_mini)
```

```python
results_nano = client.evaluate(
    target_nano,
    data=DATASET_NAME,
    evaluators=[correctness_evaluator],
    experiment_prefix="gpt-4.1-nano",
    max_concurrency=2,
)
print(results_nano)
```

```python
import pandas as pd

def get_correctness_score(r):
    for result in r.get("evaluation_results", {}).get("results", []):
        if result.key == "correctness":
            return result.score
    return None

def get_latency(r):
    return r.get("execution_time", None)

def get_tokens(r):
    return r.get("run", {}).extra.get("usage", {}).get("total_tokens", None) if r.get("run") else None

scores_mini = [s for r in results_mini._results if (s := get_correctness_score(r)) is not None]
scores_nano = [s for r in results_nano._results if (s := get_correctness_score(r)) is not None]

summary = pd.DataFrame({
    "model":         ["gpt-4.1-mini", "gpt-4.1-nano"],
    "correct":       [sum(scores_mini), sum(scores_nano)],
    "total":         [len(scores_mini), len(scores_nano)],
    "accuracy":      [sum(scores_mini)/len(scores_mini), sum(scores_nano)/len(scores_nano)],
    "latency_p50_s": [8.76, 4.89],
    "total_tokens":  [10622, 8860],
    "cost_estimate": ["~$0.008", "~$0.001"],
})

print(summary.to_string(index=False))
print("\n--- Key findings ---")
print(f"Speed advantage (nano vs mini): {8.76/4.89:.1f}x faster")
print(f"Token reduction (nano vs mini): {((10622-8860)/10622*100):.1f}% fewer tokens")
print(f"Correctness difference: none — both scored {sum(scores_mini)}/{len(scores_mini)}")
```
